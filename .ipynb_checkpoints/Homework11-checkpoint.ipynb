## Problem 1:

* Applying the forward Euler's scheme to the second order ODE:

$$
y'' + (\lambda+1)y' + \lambda y = 0, \;\; y(0) = 1, \;\; y'(0) = \lambda - 2, \;\; \lambda >> 1.
$$

Rewrite it as a system of first order ODEs. Find the region of absolute stability. (Hint: find the eigen value decomposition of the coefficient matrix, and the region of absolute stability is the domain where the step size is chosen so that each eigen value satifies the condition for the 1D ODE.)
     
   

### Solution

$$ \renewcommand{bs}{\boldsymbol}$$
Define $\bs x = \left( \begin{array} \\x_1 \\x_2 \end{array} \right) = \left( \begin{array} \\ y \\ y' \end{array} \right) $, thus $\bs x' = \left( \begin{array} \\ x_1' \\ x_2' \end{array} \right) = \left( \begin{array} \\ y' \\ y'' \end{array} \right)$. Hence, we can transform the original second order ODE into a system of first order ODEs as

$$ \left( \begin{array} \\ x_1' \\ x_2' \end{array} \right) = \left( \begin{matrix} 0 & 1 \\ -\lambda & -\lambda - 1 \end{matrix} \right) \left( \begin{array} \\x_1 \\x_2 \end{array} \right) $$

where $$ \left( \begin{array} \\x_1(0) \\x_2(0) \end{array} \right) = \left( \begin{matrix} 1 \\ \lambda - 2 \end{matrix} \right) $$

The eigenvalues of the coefficient matrix $A = \left( \begin{matrix} 0 & 1 \\ -\lambda & -\lambda - 1 \end{matrix} \right) $ is 

$$\left( \begin{matrix} \gamma_1 \\ \gamma_2 \end{matrix} \right) = \left( \begin{matrix} -1 \\ -\lambda  \end{matrix} \right)$$

Thus, for each eigenvalue, we should have the CFL condition satisfied for $\bs x' = \gamma_i \bs x$ ($i=1,2$).

The CFL condition on forward Euler requires that

$$ |1+\gamma_i h | \le 1 , i=1,2 $$

$$ \Rightarrow 0 < h \le \frac{2}{\lambda} $$.


## Problem 2:

* Build a python program based on "odeint" to solve the above ODE for different $\lambda $ values. Provide at least two cases that show different behavior of the solution based on the stiffness of the problem.
     
### Solution

Analytically, the solution can be written as

$$ y(t) = 2e^{-t} - e^{-\lambda t} $$

As can be seen, if $\lambda >> 1$, the stiffness of the problem becomes significantly larger, and the second term dominates the step size in a numerical solver.

Here the ODE system is solved numerically using ```odeint``` with $\lambda = 2$ and $\lambda = 50$, respectively. The step size $h$ is chosen to be 0.05, which satisfies the CFL condition only in the first case. Interestingly, it is found that in both cases the numerical and analytical solutions agree with each other very well. This is because ```odeint``` does not implement the forward Euler's scheme, and the time step is internally determined in order to stabilize the result. It implements several methods that can deal with both non-stiff and stiff problems, and may switch among them after checking the error after each step.

````python
lbd = 2
def f(y,t):
    return [y[1], -lbd*y[0]-(lbd+1)*y[1]]
    
y0 = [1., lbd-2]

T = 10
h = 0.05
ts = np.arange(0.0001, T, h)
ys = odeint(f, y0, ts)

plt.figure(figsize=[14,5])
plt.subplot(121)
plt.plot(ts, ys[:,:], '.-')
plt.plot(ts, 2*np.exp(-ts)-np.exp(-lbd*ts),'r')
plt.xlabel('Time',fontsize=14)
plt.title('$\lambda$ = %d and $h$ = %.2f' % (lbd,h), fontsize=14)
plt.legend(['odeint $y(t)$','odeint $y\'(t)$', 'exact $y(t)$'], fontsize=14)

lbd = 50
y0 = [1., lbd-2]
ys2 = odeint(f, y0, ts)

plt.subplot(122)
plt.plot(ts, ys2[:,:], '.-')
plt.plot(ts, 2*np.exp(-ts)-np.exp(-lbd*ts),'r')
plt.xlim(0,T/10)
plt.xlabel('Time', fontsize=14)
plt.title('$\lambda$ = %d and $h$ = %.2f' % (lbd,h), fontsize=14)
plt.legend(['odeint $y(t)$','odeint $y\'(t)$', 'exact $y(t)$'], fontsize=14)
plt.show()
````
