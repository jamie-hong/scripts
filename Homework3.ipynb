## Problem 1
$\renewcommand{bs}{\boldsymbol}$

We discussed three different ways of solving the least square problem of $\min_{\boldsymbol \beta}\Vert X \bs \beta - \boldsymbol y\Vert_2$ in the lecture: 1) the pseudo-inverse $X^+$ 2) QR decomposition 3) SVD. 

1. if $X$ is not fully ranked, i.e., its column vectors are linearly dependent, do the three methods still work?
2. if not, can you propose an modifications to the QR method, so that you can find the minimum value of $\Vert X \bs \beta - \boldsymbol y\Vert_2$ even when $X$ is rank deficient? Note that $\bs \beta$ is not unique when $X$ is not fully ranked, but there is still a deterministic answer to the minimum value of $\Vert X \bs \beta - \boldsymbol y\Vert_2$. 
3. what is an easy and effective method to find a $\bs \beta$ that is good enough when $X$ is not fully ranked in practice?

Hint:
- It is also possible to make SVD work with rank deficient matrix. Please refer to http://www.cs.cornell.edu/Courses/cs3220/2010sp/notes/svd.pdf if you are interested.

### Solution

1. If $X$ is not fully ranked, the three methods will no longer work without adjustments. For example, $ X^T X $ is no longer invertible; the pseudoinverse still exists but may not be directly computed as $(X^T X)^{-1}X$. For QR decomposition, even though it is still possible to decompose $X$ into $QR$, $R$ in this case will not be full rank and the least square solution may not be well defined. Similar problems exist for SVD.

2. Assume $ X \in \Bbb R^{m\times n}( m \ge n)$ has rank $r<n$. We first seek a decomposition of the form $XP=QR$, where $P$ is a permutation matrix chosen so that the diagnoal elements of $R$ are maximized sequentially. Then, the decomposition becomes

    $$ XP = Q \left( \begin{array} \\R & S \\ \bs 0 & \bs 0 \end{array} \right), $$
    
    where $ Q = H_1 H_2 ... H_r $ is a $m\times m$ matrix from Householder transformations, $P = P_1 P_2 ... P_r$ is a $n\times n$ permutation matrix, $R,S$ are upper triangular matrices of $r\times r$ and $r\times (n-r)$, respectively. The last $m-r$ rows are zero, because every column of $X$ is a linear combination of only the first $r$ columns of $Q$.
    
    Now, since both $Q$ and $P$ are orthonormal, we can write 
    
    $$ \begin{align}
    \Vert X \bs \beta - \bs y \Vert _2 ^2  & = \Vert Q^T(XPP^T \bs \beta - \bs y)\Vert _2 ^2 \\
    & = \left\Vert \left( \begin{array} \\R & S \\ \bs 0 & \bs 0 \end{array} \right)P^T \bs \beta - Q^T \bs y \right\Vert _2 ^2
    \end{align} $$
    
    Partition $Q^T \bs y$ as 
    
    $$ Q^T \bs y = \left( \begin{array} \\ \bs c_1 \\ \bs c_2 \\ \bs d \end{array} \right), $$
    
    where vectors $\bs c_1$, $\bs c_2$, and $\bs d$ have dimensions of $r$, $n-r$, and $m-n$, respectively.
    
    Partition $P^T \bs \beta$ as
    
    $$ P^T \bs \beta = \left( \begin{array} \\ \bs u \\ \bs v \end{array} \right),$$
    
    where $\bs u$ and $\bs v$ are vectors of dimensions $r$ and $n-r$, respectively.
    
    This yields
    
    $$ \begin{align}
    \min_{\beta} \Vert X \bs \beta - \bs y \Vert _2 ^2  & = \min_{\beta} \left \Vert \left( \begin{matrix} R \bs u + S \bs v - \bs c_1 \\ \bs c_2 \\ \bs d \end{matrix} \right) \right \Vert _2 ^2 \\
    & =  \min_{\beta} \Vert R \bs u + S \bs v - \bs c_1 \Vert _2 ^2 + \Vert \bs c_2 \Vert _2 ^2 + \Vert \bs d \Vert _2 ^2 \end{align} $$
    
    The solution becomes $ \bs u = R^{-1}(\bs c_1 - S \bs v) $ for any $ \bs v \in \Bbb R^{n-r} $.
    
    Finally, we have
    
    $$ \bs \beta^* = P P^T \bs \beta = P \left( \begin{matrix} R^{-1}(\bs c_1 - S \bs v) \\ \bs v \end{matrix} \right), $$
    
    which has infinitely many solutions because $\bs v$ is arbitrary. If we choose $\bs v = \bs 0$, then
    
    $$ \bs \beta^* = P \left( \begin{matrix} R^{-1} \bs c_1  \\ \bs 0 \end{matrix} \right) $$
    
3. Following the procedures in Part 2, we can obtain a similar solution to the least square problem using SVD. For simplification, assume $X \in \Bbb R^{n\times n}$ is a square matrix with rank $r<n$. Decomposing it as $X=U\Sigma V^T$, we have

    $$ \begin{align} 
    \left \Vert X \bs \beta - \bs y \right \Vert _2 ^2 & = \left \Vert U \Sigma V^T \bs \beta - \bs y \right \Vert _2 ^2 = \left \Vert \Sigma V^T \bs \beta - U^T \bs y \right \Vert _2 ^2 \\
    & = \left \Vert \left( \begin{matrix} \Sigma_1 & \bs 0 \\ \bs 0 & \bs 0 \end{matrix} \right) V^T \bs \beta - \left( \begin{array} \\ U_1^T \\ U_2^T \end{array} \right) \bs y \right \Vert _2 ^2 \end{align} $$
    
    Hence, analogous to the solution in Part 2, we have

    $$ \bs \beta^* = V_1 \Sigma_1^{-1} U_1^T \bs y $$
    
    Essentially, we can treat $X^+ = V_1 \Sigma_1^{-1} U_1^T$ as the pseudo inverse for the rank-deficient matrix $X$, so that the solution to the problem $X \bs \beta - \bs y \approx 0$ can be approximated as $\bs \beta = X^+ \bs y$.

### Solution 

1. The L-2 condition number of the covariance matrix is 74717.1.
2. The singular values of $V$ is listed in the table below. The ratio between the largest and smallest singluar values is 74717.1, which is the same as the L-2 norm of $V$.
3. There could be problems when computing $\bs y = V^{-1} \bs x$ as the L-2 norm of $V^{-1}$ is very large. As shown below, the 4th entry of $\bs x$ is changed from 1 to 1.005, the resulting $y$ is considerably different from the previous result, especially for the fourth entry.

```python
V = cmt_rates.cov()
V_cond = cond(V,2)
print("Condition number: %.1f" % cond(V,2))

u,s,v= svd(V)
fmt.displayDF(pd.DataFrame(s, columns=["Singular Values of V"]).T, fmt="3g")
print("\nRatio between first and last singular values: %.1f" % (s[0]/s[-1]))

VV = sp.MatrixSymbol('V', 9, 9)
VI = sp.MatrixSymbol('V^(-1)',9,9)
vi = sp.Matrix(inv(V))
x = sp.Matrix([1,1,1,1,1,1,1,1,1])
xp = sp.Matrix([1,1,1,1.005,1,1,1,1,1])
print ("\nBefore perturbation: ")
fmt.displayMath(vi.evalf(3),fmt.joinMath('=', x, sp.Matrix(vi*x).evalf(3)), sep="", pre="\\scriptsize")
print ("\nAfter perturbation: ")
fmt.displayMath(vi.evalf(3),fmt.joinMath('=', xp, sp.Matrix(vi*xp).evalf(3)), sep="", pre="\\scriptsize")
```

### Solution

**1)** The covariance matrix of the daily interest rate changes is displayed below:

```python
delta = cmt_rates.diff()[1:] # rate changes
V_d = delta.cov()
fmt.displayDFs(V_d, headers=["Covariance Matrix of Daily Interest Rate Changes"],fmt="4f")
```

**2)** PCA on the correlation ($C_d$) and covariance ($V_d$) matrices of daily rate changes give different principal components, as PCA is scale variant. The correlation matrix is equivalent to the "covariance" matrix of standardized data (centered and rescaled), which is essentially a different data set. Thus, it is natural that the PCA of the covariance and correlation matrices yield different principal components.

Each eigenvector after EVD has already been normalized with a unit norm, thus we can directly compare these vectors. It can be easily seen that none of the eigenvectors of $V_d$ and $C_d$ are equivalent.

```python
C_d = delta.corr()
ev_V, evec_V = eig(V_d)
ev_C, evec_C = eig(C_d)
fmt.displayDFs(pd.DataFrame(evec_V, columns=["PC %d" % x for x in range(1,10,1)],index=tenors), 
               pd.DataFrame(evec_C, columns=["PC %d" % x for x in range(1,10,1)],index=tenors), 
               headers=["Eigenvectors of Covariance Matrix", "Eigenvectors of Correlation Matrix"],fmt="3g")
```

**3)** Using the covariance matrix, 3 principal components are needed to explain 95% of the variance; using the correlation matrix, 3 principal components are needed as well.

```python
pct_V = np.cumsum(ev_V)/sum(ev_V)*100
fmt.displayDF(pd.DataFrame({'Covariance PC':range(1, len(ev_V)+1), 
                            'Eigenvalues':ev_V, 'Cumulative Var (%)': pct_V}).set_index(['Covariance PC']).T, "4f")

pct_C = np.cumsum(ev_C)/sum(ev_C)*100
fmt.displayDF(pd.DataFrame({'Correlation PC':range(1, len(ev_C)+1), 
                            'Eigenvalues':ev_C, 'Cumulative Var (%)': pct_C}).set_index(['Correlation PC']).T, "4f")
```

**4)** The left figure below plots the first three principal components/factors. Factor 1 places roughly the same weights on the rate changes across the tenors (with slighly more on the middle-maturity rates), which means that the daily changes in rates across all tenors at a point in time primarily depends on the average change of the rates. This is similar to the effect of "shift" in the level of rates as illustrated in the lecture. Factor 2 places large positive weights on the shorter tenors and negative (but smaller) weights on the longer ones, which means this factor examines the impact on rate changes due to changes in one end of the term structure. This is similar to the "slope" change of interest rate levels. Finally, Factor 3 places large positive weights on the short and long maturities and large but negative weights on the middle ones, which means this factor examines the impact on rate changes due to the "curvature" of the rate change schedule.  

```python
flab = ['factor %d' % i for i in range(1, 4)]
tenors = cmt_rates.columns.map(float)
fig = figure(figsize=[14, 5])
ax1 = fig.add_subplot(121)
plot(tenors, evec_V[:, :3], '.-');
xlabel('Tenors(Y)')
legend(flab, loc='best')
title('First 3 Principal Components');

fs = delta.dot(evec_V).iloc[:, :3]

ax2 = fig.add_subplot(122)
fs.plot(ax=ax2, title='Projections to the First 3 PCs')
legend(flab, loc='best');
```

**5)** The history of PCA scores, i.e., the projections of the history data to the first 3 principal components, are plotted in the right figure above. As can be seen, the first factor (blue) explains most of the variations of rate changes in the period of data, and the variations explained by Factors 2 and 3 are significantly smaller.

**6)** Assuming the interest rate changes are independent normals with mean 0, we use the first 3 principal components to simulate 20 years of rate changes (5,000 data points). 

It is found that the covariance matrix of the simulated data is similar to that of the historical data. Their differences can be represented by matrix norms of $V_d - V_{ds}$, where $V_d$ is the covariance matrix of *historical* rate changes, and $V_{ds}$ is the covariance matrix of *simulated* rate changes.

```python
H_V = np.diag(ev_V[0:3]**0.5)
L_V = evec_V[:,:3].dot(H_V)
Z = np.random.randn(3,5000)
dt = 1
Sdw = L_V.dot(Z)*np.sqrt(dt) # simulated rate changes
df = pd.DataFrame(Sdw.T, columns=tenors)
V_ds = df.cov() # covariance matrix of simulated data
fmt.displayDFs(V_ds,headers=["Covariance Matrix of Simulated Daily Rates Change"],fmt="3g")

VDiff = np.mat(V_d)-np.mat(V_ds)
fmt.displayDF(pd.DataFrame([norm(VDiff, n) for n in (1, 2, inf)],
             index = ["d-1", "d-2", "d-$\infty$"], columns=['Distance']), 
              fmt="4g")
```

**7)** The starting interest rates in the simulated period should be the ending values of the historical data.

It can be seen that without a specific stochastic model, the interest rates may go below 0 because we assumed the changes in rates are independent normals. 

```python
ending = np.array(cmt_rates.iloc[-1,:])
fmt.displayDF(pd.DataFrame(ending, index=tenors, columns=["Ending Historical Rates"]).T,fmt="4f")
SdwT = np.array(Sdw.T)
SdwT[0,:] = Sdw.T[0,:] + ending
SimRates = np.cumsum(SdwT,0)  # simulated rates
df_s = pd.DataFrame(SimRates, columns=tenors)
ax = df_s.plot(figsize=[11,8])
ax.set_title("Simulated Interest Rates",fontsize=18)
ax.set_xlabel("Time(Year)",fontsize=18)
ax.set_xticklabels(np.arange(0,21,4))
ax.set_ylabel("Rate(%)",fontsize=18)
ax.legend(loc="upper left")
fmt.displayDFs(pd.DataFrame({'Tenor':tenors,
                            'Mean':df_s.mean(), 
                            'Standard Deviation': df_s.std(),
                            '2% Quantile': df_s.quantile(q=0.02),
                            '98% Quantile': df_s.quantile(q=0.98)}).set_index(['Tenor']).T, 
                            headers=["Statistics of Simulated Interest Rate Levels"], fmt="4f")
```
